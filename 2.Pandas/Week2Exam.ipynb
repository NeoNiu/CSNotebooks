### 1.学生数据分析（总分数：60）

#### 1.数据导入并展示, 文件位置位于0.Teacher/data/下（10分）
   - 请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（5分）   
   - 请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（5分）

```python
import pandas as pd
df_info = pd.read_csv('student-info.csv',sep=';')
df_info.head()
```

```python
df_score = pd.read_csv('student-score.csv',sep=';')
df_score.head()
```

#### 2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (5分）

```python
df = pd.merge(df_info, df_score, on = 'ID')
df.head()
```

#### 3.把"df"所有的列展示出来(5分)

```python
df.columns
```

#### 4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
sex_score = df.groupby('sex')['G1', 'G2', 'G3'].agg('mean')
sex_score
```

#### 5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
age_score = df.groupby('age')['G1','G2','G3'].agg('mean')
age_score.plot(kind = 'bar')
```

#### 6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
school_score = df.groupby('school')['G1','G2','G3'].mean()
school_score.T.head()
school_score.T.plot.bar()
```

#### 7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(10分)

```python
def top10(item):
    return item.loc[item['G'].nlargest(10).index]
df['G'] = df['G1']+df['G2']+df['G3']
df.head()
df.groupby('school').apply(top10)
```

### 2.沃尔玛销售数据整理（共40分）

#### 1. 数据导入并展示, 文件位置位于0.Teacher/data/walmart下 （10分）

   - 请将文件 stores.csv 中的内容读入pandas DataFrame "df_stores"，并展示前五行以及打印df_stores的shape （2分）

```python
df_stores = pd.read_csv("stores.csv")
df_stores.head()
```

```python
df_stores.shape
```

   - 请将文件 features.csv 中的内容读入pandas DataFrame "df_feats"，将df_feats改成只留下以下几个columns: "Store", "Date", "Temperature", "Fuel_Price", "CPI", "Unemployment", 并展示前五行以及打印df_feats的shape  （4分）

```python
df_feats = pd.read_csv("features.csv", parse_dates=["Date"])
df_feats = df_feats[['Store', 'Date', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']]
df_feats.head()
```

```python
df_feats.shape
```

   - 请将文件 sales.csv 中的内容读入pandas DataFrame "df_sales"，并展示前五行以及打印df_sales的shape（4分）

```python
df_sales = pd.read_csv("sales.csv" , parse_dates=['Date'])
df_sales.head()
```

```python
df_sales.shape
```

#### 8. 我们发现 df_sales 中每个Store有很多个Dept(department)， 我们只关心整家店的销售情况，所以请把这张表格按照Store和Date来计算整店销售总额，IsHoliday这一列不必保留。结果保存在df_sales_by_store中。（10分）

```python
df_sales_by_store = df_sales.groupby(['Store','Date']).agg({"Weekly_Sales":'sum'})
df_sales_by_store.head()
```

```python
df_sales_by_store = df_sales[["Store", "Date", "Weekly_Sales", \
                     "IsHoliday"]].groupby(["Store", "Date"]).agg({"Weekly_Sales": sum})
df_sales_by_store.head()
```

```python
pd.__version__
```

#### 9. 使用 Store 和 Date 这两列合并 df_sales_by_store 和 df_feats 两张表格，结果保存在 df_sale_feats 。（10分）

```python
df_sales_feats  = pd.merge(left = df_sales_by_store, right = df_feats, on =['Store','Date'])
df_sales_feats.head()
```

```python
df_sale_feats = pd.merge(df_sales_by_store, df_feats, left_on=["Store", "Date"], right_on=["Store", "Date"])
df_sale_feats.head()
```

#### 10. 使用 Store 这一列合并 df_sale_feats 和 df_stores 两张表格，结果仍然保留在df_sale_feats 中。（10分）

```python
df_sale_feats = pd.merge(df_sale_feats, df_stores, on=["Store"])
df_sale_feats.head()
```
